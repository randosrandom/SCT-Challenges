```python
import pandas as pd
```

```python
# Import predictions

date = '2022_05_04_SubmissionScripts'

pred_Lazio = pd.read_csv("/content/drive/Othercomputers/Laptop/homework3/scripts/Submissions/" + date + '/webeepUpload_Lazio.csv')
pred_Lombardia = pd.read_csv("/content/drive/Othercomputers/Laptop/homework3/scripts/Submissions/" + date + '/webeepUpload_Lombardia.csv')
pred_Sicilia = pd.read_csv("/content/drive/Othercomputers/Laptop/homework3/scripts/Submissions/" + date + '/webeepUpload_Sicilia.csv')

pred_Lazio = pred_Lazio.iloc[:, 1:]
pred_Lombardia = pred_Lombardia.iloc[:, 1:]
pred_Sicilia = pred_Sicilia.iloc[:, 1:]
```

```python
# Import true data

file_path = "/content/drive/Othercomputers/Laptop/homework3/Data/dpc-covid19-ita-regioni.csv"
features_list = ["totale_ospedalizzati", "nuovi_positivi", "dimessi_guariti", "deceduti"]

# Load data
df_all = pd.read_csv(
    file_path,
    usecols = ["data", "denominazione_regione", "nuovi_positivi", "totale_ospedalizzati", "dimessi_guariti", "deceduti"],
    parse_dates = True,
)

# Convert data coulumn
df_all["data"] = pd.to_datetime(df_all["data"]).dt.date
```

```python
df_tot = pd.DataFrame()
    
df_by_region = dict() # dictionary containing data for each regione

regions = list(df_all.loc[:,"denominazione_regione"].unique())

for reg in regions:
    
    df_by_region[reg] = df_all.loc[df_all["denominazione_regione"] == reg]
    columns_={"totale_ospedalizzati": "totale_ospedalizzati_" + reg, 
              "nuovi_positivi": "nuovi_positivi_" + reg, 
              "dimessi_guariti" : "dimessi_guariti_" + reg, 
              "deceduti": "deceduti_" + reg}
    df_by_region[reg].pop("denominazione_regione")
    df_by_region[reg] = df_by_region[reg].rename(columns = columns_)
    df_by_region[reg].set_index("data", inplace = True)
    
    if df_tot.shape[0] == 0:
        df_tot = df_by_region[reg].iloc[:,1:]
    else:
        df_tot = pd.concat([df_tot, df_by_region[reg].iloc[:,1:]], axis=1)
```

```python
true_Lazio = df_by_region['Lazio'].iloc[-1,:]
true_Lombardia = df_by_region['Lombardia'].iloc[-1,:]
true_Sicilia = df_by_region['Sicilia'].iloc[-1,:]
```

# Lazio

```python
reg = 'Lazio'

true_Lazio = pd.DataFrame(true_Lazio).T

pred_Lazio.index = true_Lazio.index

original_concatenated_df_Lazio = pd.concat([pred_Lazio, true_Lazio], axis=1 )

original_errors_Lazio = pd.DataFrame()
for i in range(4):
  original_errors_Lazio = pd.concat([original_errors_Lazio, original_concatenated_df_Lazio.iloc[:, i] - original_concatenated_df_Lazio.iloc[:, i+4]], axis = 1, ignore_index=True)

original_err_Lazio = original_errors_Lazio.rename(columns={0: 'totale_ospedalizzati_' + reg, 
                                                           1: 'nuovi_positivi_' + reg,
                                                           2: 'dimessi_guariti_' + reg,
                                                           3: 'deceduti_' + reg} )


original_error_perc_Lazio = (original_err_Lazio / true_Lazio) * 100
```
